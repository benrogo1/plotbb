```python
import pandas as pd
import numpy as np
import plotly.express as px

data = pd.read_csv('trait_data_mammals.csv')

data['log_adult_mass_g'] = np.log(data['adult_mass_g'])
data['log_brain_mass_g'] = np.log(data['brain_mass_g'])

fig = px.scatter(data, x='log_brain_mass_g', y='log_adult_mass_g', color='order', 
                 color_discrete_sequence=px.colors.qualitative.Bold, 
                 hover_data=['species'], 
                 title='Adult Mass vs Brain Mass')
fig.update_layout(xaxis_title='Log Brain Mass (g)', yaxis_title='Log Adult Mass (g)', hovermode='closest')

# Save as HTML
fig.write_html("interactive_plot.html")
fig.show()
```
